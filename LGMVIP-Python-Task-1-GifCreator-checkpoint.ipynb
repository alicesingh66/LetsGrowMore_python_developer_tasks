```python
from PIL import Image, ImageDraw

# Create a list to hold the frames
frames = []

# Create 10 sample frames with different colors
for i in range(10):
    # Create a new image with white background
    img = Image.new('RGB', (200, 200), 'white')
    draw = ImageDraw.Draw(img)
    
    # Draw a rectangle with varying colors
    draw.rectangle([50, 50, 150, 150], fill=(i*25, i*15, i*5))
    
    # Append image to the frame list
    frames.append(img)
```

```python
# Save the frames as a GIF
frames[0].save('/mnt/data/animated.gif', save_all=True, append_images=frames[1:], optimize=False, duration=100, loop=0)
```

```python
from PIL import Image, ImageDraw
from IPython.display import Image as IPImage, display

# Create a list to hold the frames
frames = []

# Create 10 sample frames with different colors
for i in range(10):
    # Create a new image with white background
    img = Image.new('RGB', (200, 200), 'white')
    draw = ImageDraw.Draw(img)
    
    # Draw a rectangle with varying colors
    draw.rectangle([50, 50, 150, 150], fill=(i*25, i*15, i*5))
    
    # Append image to the frame list
    frames.append(img)

# Save the frames as a GIF in the current working directory
gif_path = 'animated.gif'
frames[0].save(gif_path, save_all=True, append_images=frames[1:], optimize=False, duration=100, loop=0)

# Display the created GIF
display(IPImage(filename=gif_path))
```
